# Part A : Data Preprocessing

# import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# import dataset

```python
dataset = pd.read_csv('Logistic Data.csv')
```

```python
dataset
```

# create feature matrix X and dependent vector Y

```python
X= dataset.iloc[:,:-1].values
Y= dataset.iloc[:,-1].values
```

```python
X
```

```python
Y
```

# replacing missing data

```python
from sklearn.impute import SimpleImputer
```

```python
imputer= SimpleImputer(missing_values=np.nan,strategy='mean')
X= imputer.fit_transform(X)
```

```python
X
```

# encoding not reqd

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc= StandardScaler()
X= sc.fit_transform(X)
```

```python
X
```

# spliting the dataset into training and testing

```python
from sklearn.model_selection import train_test_split
```

```python
Xtrain,Xtest,Ytrain,Ytest= train_test_split(X,Y,test_size=0.2,random_state=1)
```

# Part B : Building the Machine Learning model

# training the Logistic Regression model

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr= LogisticRegression(random_state=0)
lr.fit(Xtrain,Ytrain)
```

# testing the model

```python
Yestimated= lr.predict(Xtest)
```

```python
Yestimated
```

```python
Ytest
```

```python
np.concatenate((Yestimated.reshape(len(Yestimated),1),Ytest.reshape(len(Ytest),1)),1)
```

# Part C : Performance Metric

```python
# Confusion matrix : used to evaluate the performance of a classification model
from sklearn.metrics import confusion_matrix
cm= confusion_matrix(Ytest,Yestimated)
print('Confusion matrix : ')
print(cm)
```

```python
from sklearn.metrics import accuracy_score,precision_score
ac= accuracy_score(Ytest,Yestimated)
ps= precision_score(Ytest,Yestimated)
print('Accuracy score: ',ac)
print('Precision score: ',ps)
```
